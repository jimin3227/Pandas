## concat()

```python
s2 = pd.Series(['a','b'], index=[3,4])
s1 = pd.Series(['c','d'], index=[1,2])
pd.concat([s1,s2])
```

```python
def create_df(cols, idx):
    data = {c: [str(c.lower()) + str(i) for i in idx] for c in cols}
    return pd.DataFrame(data, idx)
```

```python
df1 = create_df('AB', [1,2])
df1
```

```python
df2 = create_df('AB', [3,4])
df2
```

```python
pd.concat([df1,df2])
```

```python
df3 = create_df('AB', [0,1])
df3
```

```python
df4 = create_df('CD', [0,1])
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4], axis=1)
```

```python
# pd.concat([df1,df3], verify_integrity=True) # verify_integrity=True: 잘못된 concat이 일어나지 않도록 방지
```

```python
pd.concat([df1,df3], ignore_index=True)
```

```python
pd.concat([df1,df3], keys=['X','Y'])
```

```python
df5 = create_df('ABC', [1,2])
df6 = create_df('BCD', [3,4])
pd.concat([df5,df6])
```

```python
pd.concat([df5,df6], join='inner') # 교집합만 출력
```
